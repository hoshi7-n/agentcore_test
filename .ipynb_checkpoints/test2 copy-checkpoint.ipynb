## æ¤œè¨¼ã‚·ãƒŠãƒªã‚ª

2.  **Episode 2 (ç¾åœ¨)**: ãƒ¦ãƒ¼ã‚¶ãƒ¼ã€Œã¾ãŸLambdaãŒå‹•ã‹ãªã„ã‚“ã ã‚ˆã­ã€œã€

```python
# 1. å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import uuid
import time
import json
import boto3
from datetime import datetime

try:
    from bedrock_agentcore.memory import MemoryClient
    print("Successfully imported MemoryClient.")
except ImportError:
    print("Installing bedrock-agentcore...")
    !pip install bedrock-agentcore strands-agents strands-agents-tools --quiet
    from bedrock_agentcore.memory import MemoryClient

# ã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆåˆæœŸåŒ–
REGION = "us-east-1" 
memory_client = MemoryClient(region_name=REGION)
bedrock_runtime = boto3.client('bedrock-runtime', region_name=REGION)

ACTOR_ID = "engineer_tanaka_001"
```

## 2. ãƒ¡ãƒ¢ãƒªãƒªã‚½ãƒ¼ã‚¹ä½œæˆé–¢æ•°
Episodeæ©Ÿèƒ½ã®ON/OFFã‚’åˆ‡ã‚Šæ›¿ãˆã¦ãƒ¡ãƒ¢ãƒªã‚’ä½œæˆã—ã¾ã™ã€‚

```python
def create_memory_resource(enable_episode=False):
    unique_id = str(uuid.uuid4())[:6]
    memory_name = f"trouble_mem_{unique_id}"
    
    # Semantic Memory (äº‹å®Ÿã®è¨˜æ†¶)
    strategies = [
        {
            "semanticMemoryStrategy": {
                "name": "factStrategy",
                "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}/facts"]
            }
        }
    ]
    
    # Episodeæ©Ÿèƒ½ã®ON/OFF
    if enable_episode:
        strategies.append({
            "episodicMemoryStrategy": {
                "name": "episodeStrategy",
                # ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰ï¼ˆã‚»ãƒƒã‚·ãƒ§ãƒ³ã”ã¨ã®è¦ç´„ï¼‰
                "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}/sessions/{sessionId}"],
                # Reflectionï¼ˆæ´å¯Ÿï¼‰: ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰ã®è¦ªéšå±¤ã«è¨­å®šã™ã‚‹å¿…è¦ã‚ã‚Š
                "reflectionConfiguration": {
                     "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}"]
                }
            }
        })
        label = "[Episode ON]"
    else:
        label = "[Episode OFF]"

    print(f"Creating Memory {label}: {memory_name} ...")
    memory_info = memory_client.create_memory_and_wait(
        name=memory_name,
        description=f"Troubleshooting memory {label}",
        strategies=strategies
    )
    return memory_info['id']

# 2ã¤ã®ãƒ¡ãƒ¢ãƒªã‚’ä½œæˆ
mem_id_off = create_memory_resource(enable_episode=False)
mem_id_on = create_memory_resource(enable_episode=True)
```

## 3. éå»ã®æ—¥å ±ãƒ‡ãƒ¼ã‚¿ã®æŠ•å…¥
ã€Œç’°å¢ƒå¤‰æ•°å¿˜ã‚Œã€ã®å¤±æ•—è«‡ã‚’æ³¨å…¥ã—ã¾ã™ã€‚

```python
def inject_trouble_reports(memory_id):
    # --- Episode 1: éå»ã®å¤±æ•— ---
    session_id_1 = f"daily_report_20241001_{str(uuid.uuid4())[:4]}"
    messages_1 = [
        ("ä»Šæ—¥ã®æ—¥å ±ã§ã™ã€‚AWS Lambdaã®å®Ÿè£…ã§3æ™‚é–“ã‚‚ãƒãƒã£ã¦ã—ã¾ã„ã¾ã—ãŸã€‚", "USER"),
        ("ãã‚Œã¯å¤§å¤‰ã§ã—ãŸã­ã€‚ã©ã®ã‚ˆã†ãªã‚¨ãƒ©ãƒ¼ã§ã—ãŸã‹ï¼Ÿ", "ASSISTANT"),
        ("å®Ÿè¡Œæ™‚ã«ã‚­ãƒ¼ã‚¨ãƒ©ãƒ¼ãŒå‡ºã¦å‹•ã‹ãªã‹ã£ãŸã‚“ã§ã™ã€‚ã‚³ãƒ¼ãƒ‰ã¯åˆã£ã¦ã‚‹ã¯ãšãªã®ã«ã€‚", "USER"),
        ("åŸå› ã¯ç‰¹å®šã§ãã¾ã—ãŸã‹ï¼Ÿ", "ASSISTANT"),
        ("ã¯ã„ã€‚çµå±€ã€Lambdaã®ã‚³ãƒ³ã‚½ãƒ¼ãƒ«ç”»é¢ã§ç’°å¢ƒå¤‰æ•° `DB_CONNECTION_STRING` ã‚’è¨­å®šã—å¿˜ã‚Œã¦ãŸã®ãŒåŸå› ã§ã—ãŸã€‚", "USER"),
        ("ãªã‚‹ã»ã©ã€ç’°å¢ƒå¤‰æ•°ã®è¨­å®šæ¼ã‚Œã¯ã‚ˆãã‚ã‚‹ãƒŸã‚¹ã§ã™ã­ã€‚è¨˜éŒ²ã—ã¦ãŠãã¾ã™ã€‚", "ASSISTANT")
    ]
    
    print(f"Injecting Episode 1 (Env Var Failure) -> {memory_id} ...")
    memory_client.create_event(
        memory_id=memory_id,
        actor_id=ACTOR_ID,
        session_id=session_id_1,
        messages=messages_1
    )
    print("Done.")

print("--- Processing Episode OFF Memory ---")
inject_trouble_reports(mem_id_off)

print("\n--- Processing Episode ON Memory ---")
inject_trouble_reports(mem_id_on)
```

## 4. è¨˜æ†¶ã®å–å¾—ã¨ã‚¨ãƒ¼ã‚¸ã‚§ãƒ³ãƒˆå›ç­”ï¼ˆä¿®æ­£ç‰ˆï¼‰

ã“ã“ã§ `list_items` ã‚’ä½¿ç”¨ã—ã¦ã€æ­£ã—ã„Namespaceã‹ã‚‰è¨˜æ†¶ã‚’å–å¾—ã—ã¾ã™ã€‚
å‡¦ç†å®Œäº†ã¾ã§ç´„60ç§’å¾…ã¡ã¾ã™...

```python
print("Waiting for AgentCore processing (sleeping 60s)...")
time.sleep(60)
agent_runtime_client = boto3.client('bedrock-agent-runtime', region_name=REGION)

def get_memory_context(memory_id):
    """
    ãƒ¡ãƒ¢ãƒªå†…ã®æƒ…å ±ã‚’å–å¾—ã—ã¦ãƒ†ã‚­ã‚¹ãƒˆåŒ–ã™ã‚‹ãƒ˜ãƒ«ãƒ‘ãƒ¼é–¢æ•°
    (ä¿®æ­£ç‰ˆ: Boto3ç›´æ¥åˆ©ç”¨ + Strategyåˆ¤å®šãƒ­ã‚¸ãƒƒã‚¯å¼·åŒ–)
    """
    context_text = ""
    
    try:
        # 1. ãƒ¡ãƒ¢ãƒªè¨­å®šã®å–å¾— (ã“ã“ã¯ MemoryClient ã§ã‚‚OKã§ã™ãŒã€Boto3ã§ã‚‚å–ã‚Œã¾ã™)
        # æ—¢å­˜ã®ã‚³ãƒ¼ãƒ‰ã‚’ç”Ÿã‹ã™ãŸã‚ memory_client ã‚’ä½¿ç”¨
        mem_details = memory_client.get_memory(memoryId=memory_id)
        
        memory_data = mem_details.get('memory', {})
        strategies = memory_data.get('strategies', [])
        
        target_strategy_id = None
        
        # Strategyã®ç‰¹å®š (name="episodeStrategy" ã‚’æ¢ã™)
        for s in strategies:
            if s.get('name') == 'episodeStrategy' or s.get('type') == 'episodic':
                target_strategy_id = s.get('strategyId')
                break 

        # --- A. Episode StrategyãŒã‚ã‚‹å ´åˆ ---
        if target_strategy_id:
            # ã‚»ãƒƒã‚·ãƒ§ãƒ³ä¸€è¦§ã‚’å–å¾—ã™ã‚‹ãŸã‚ã®Namespace
            namespace = f"/strategies/{target_strategy_id}/actors/{ACTOR_ID}/sessions"
            
            print(f"DEBUG: Fetching items via Boto3 from: {namespace}")

            try:
                # â˜…ä¿®æ­£ãƒã‚¤ãƒ³ãƒˆ: memory_client ã§ã¯ãªã agent_runtime_client ã‚’ä½¿ç”¨
                response = agent_runtime_client.list_items(
                    memoryId=memory_id,
                    namespace=namespace,
                    maxResults=10  # Boto3ã®ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã¯CamelCaseã®å ´åˆãŒå¤šã„ã®ã§æ³¨æ„
                )
                
                items = response.get('items', [])
                
                if items:
                    print(f"  [Episode ON] {len(items)}å€‹ã®ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰(è¦ç´„)ãŒè¦‹ã¤ã‹ã‚Šã¾ã—ãŸ")
                    for item in items:
                        # æ§‹é€ : item['content'] -> {'summary': '...', 'text': '...'}
                        content = item.get('content', {}) 
                        
                        # è¦ç´„(summary) ã¾ãŸã¯ ãƒ†ã‚­ã‚¹ãƒˆ(text) ã‚’å–å¾—
                        # å¤±æ•—è«‡ã®è¦ç´„ãŒå…¥ã£ã¦ã„ã‚‹ã¯ãšã§ã™
                        summary = content.get('summary') or content.get('text')
                        
                        if summary:
                            context_text += f"- éå»ã®è¨˜æ†¶: {summary}\n"
                else:
                    print("  [Episode ON] ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰ã¯è¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã§ã—ãŸ (0ä»¶)")
                    
            except Exception as e:
                print(f"  [Warning] list_items failed: {e}")
                # è©³ç´°ãƒ‡ãƒãƒƒã‚°ãŒå¿…è¦ãªå ´åˆã¯ã‚³ãƒ¡ãƒ³ãƒˆã‚¢ã‚¦ãƒˆã‚’å¤–ã™
                # import traceback
                # traceback.print_exc()
                
        else:
            print("  [Warning] Episode Strategyã®è¨­å®šãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã§ã—ãŸ (åˆ¤å®šãƒ­ã‚¸ãƒƒã‚¯ã§ãƒ’ãƒƒãƒˆãªã—)")

    except Exception as e:
        print(f"  Error fetching memory: {e}")
        
    if not context_text:
        context_text = "(ç‰¹ã«é–¢é€£ã™ã‚‹è¨˜æ†¶ã¯ã‚ã‚Šã¾ã›ã‚“)"
        
    return context_text

def ask_agent(memory_id, user_query, title):
    print(f"\n{'='*60}")
    print(f" {title}")
    print(f"{'='*60}")

    # 1. è¨˜æ†¶ã‚’å–å¾—
    context = get_memory_context(memory_id)
    
    # 2. ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆä½œæˆ
    prompt = f"""
ã‚ãªãŸã¯å„ªç§€ãªã‚·ãƒ‹ã‚¢ã‚¨ãƒ³ã‚¸ãƒ‹ã‚¢ã®ã‚¢ã‚·ã‚¹ã‚¿ãƒ³ãƒˆã§ã™ã€‚
ä»¥ä¸‹ã®ã€éå»ã®è¨˜æ†¶ã€‘ã‚’å‚è€ƒã«ã—ã¦ã€ãƒ¦ãƒ¼ã‚¶ãƒ¼ã®å›°ã‚Šã”ã¨ã«å¯¾ã—ã¦ã‚¢ãƒ‰ãƒã‚¤ã‚¹ã‚’ã—ã¦ãã ã•ã„ã€‚
ã‚‚ã—è¨˜æ†¶ã®ä¸­ã«ã€ä»Šå›ã®ãƒˆãƒ©ãƒ–ãƒ«ã®åŸå› ã«ãªã‚Šãã†ãªã€Œéå»ã®å¤±æ•—ãƒ‘ã‚¿ãƒ¼ãƒ³ã€ãŒã‚ã‚Œã°ã€ãã‚Œã‚’å…·ä½“çš„ã«æŒ‡æ‘˜ã—ã¦ãã ã•ã„ã€‚

ã€éå»ã®è¨˜æ†¶ã€‘
{context}

ãƒ¦ãƒ¼ã‚¶ãƒ¼: {user_query}
AI:"""

    # 3. Amazon Nova Micro ã§å›ç­”ç”Ÿæˆ
    try:
        model_id = "amazon.nova-micro-v1:0"
        body = json.dumps({
            "inferenceConfig": {"max_new_tokens": 512},
            "messages": [
                {"role": "user", "content": [{"text": prompt}]}
            ]
        })

        response = bedrock_runtime.invoke_model(modelId=model_id, body=body)
        response_body = json.loads(response.get('body').read())
        ai_response = response_body['output']['message']['content'][0]['text']
        
        print(f"\n  [ğŸ¤– AIã‚¨ãƒ¼ã‚¸ã‚§ãƒ³ãƒˆã®å›ç­” (Nova Micro)]")
        print(f"  {ai_response}")
        
    except Exception as e:
        print(f"  Invoke Error: {e}")


# æ¤œè¨¼å®Ÿè¡Œï¼
query = "LambdaãŒå‹•ã‹ãªã„ã‚“ã ã‚ˆã­ã€œãªã‚“ã§ã ã‚ã†"

ask_agent(mem_id_off, query, "âŒ Episode OFF ã®å ´åˆ")
ask_agent(mem_id_on, query, "âœ… Episode ON ã®å ´åˆ")
```

```python
# ã‚¯ãƒªãƒ¼ãƒ³ã‚¢ãƒƒãƒ—
memory_client.delete_memory(memory_id=mem_id_off)
memory_client.delete_memory(memory_id=mem_id_on)
print("Cleanup complete.")
```
