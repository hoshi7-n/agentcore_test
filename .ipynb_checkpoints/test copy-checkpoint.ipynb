## ä¿®æ­£å†…å®¹
å…ƒã®ã‚³ãƒ¼ãƒ‰ã§ã¯ `bedrock_agent_runtime.retrieve` (Knowledge Baseç”¨API) ã‚’ä½¿ç”¨ã—ã¦ã„ã¾ã—ãŸãŒã€AgentCore Memoryã¯å°‚ç”¨ã®APIã§æ¤œç´¢ã™ã‚‹å¿…è¦ãŒã‚ã‚‹ãŸã‚ã€`bedrock_agentcore` ã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆã® `retrieve_memory_records` ã‚’ä½¿ç”¨ã™ã‚‹ã‚ˆã†ã«ä¿®æ­£ã—ã¾ã—ãŸã€‚

## æ¤œè¨¼ã‚·ãƒŠãƒªã‚ª

    * ãƒ¦ãƒ¼ã‚¶ãƒ¼: ã€ŒLambdaãŒå‹•ã‹ãªãã¦ãƒãƒã£ãŸã€‚åŸå› ã¯ **ç’°å¢ƒå¤‰æ•°ã®è¨­å®šæ¼ã‚Œ** ã ã£ãŸã€‚ã€
2.  **Episode 2 (ç¾åœ¨)**:
    * ãƒ¦ãƒ¼ã‚¶ãƒ¼: ã€Œã¾ãŸLambdaãŒå‹•ã‹ãªã„ã‚“ã ã‚ˆã­ã€œãªã‚“ã§ã ã‚ã†ã€

## æ¯”è¼ƒãƒã‚¤ãƒ³ãƒˆ
* **Episode OFF**: éå»ã®æ–‡è„ˆã‚’æ€ã„å‡ºã›ãšã€ã€Œãƒ­ã‚°ã‚’ç¢ºèªã—ã¦ãã ã•ã„ã€ã€Œæ¨©é™ã¯ã©ã†ã§ã™ã‹ã€ã¨ã„ã£ãŸ**ä¸€èˆ¬çš„ãªå›ç­”**ã«ãªã‚‹ã€‚
* **Episode ON**: éå»ã®ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰ï¼ˆç’°å¢ƒå¤‰æ•°å¿˜ã‚Œï¼‰ã‚’æ€ã„å‡ºã—ã€ã€Œ**å‰ã¨åŒã˜ã‚ˆã†ã«ã€ç’°å¢ƒå¤‰æ•°ã‚’å¿˜ã‚Œã¦ã„ã¾ã›ã‚“ã‹ï¼Ÿ**ã€ã¨**æ–‡è„ˆã«æ²¿ã£ãŸé‹­ã„æŒ‡æ‘˜**ãŒã§ãã‚‹ã€‚

```python
# 1. å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import uuid
import time
import json
import boto3
from datetime import datetime

try:
    from bedrock_agentcore.memory import MemoryClient
    print("Successfully imported MemoryClient.")
except ImportError:
    print("Installing bedrock-agentcore...")
    !pip install bedrock-agentcore strands-agents strands-agents-tools --quiet
    from bedrock_agentcore.memory import MemoryClient

# ã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆåˆæœŸåŒ–
REGION = "us-east-1" 

# Memoryæ“ä½œç”¨ (AgentCore SDK)
memory_client = MemoryClient(region_name=REGION)

# AgentCoreãƒ‡ãƒ¼ã‚¿æ¤œç´¢ç”¨ (Boto3)
# â€» ã“ã“ãŒä¿®æ­£ãƒã‚¤ãƒ³ãƒˆ: KBç”¨ã® bedrock-agent-runtime ã§ã¯ãªã bedrock-agentcore ã‚’ä½¿ç”¨ã—ã¾ã™
try:
    agentcore_runtime = boto3.client('bedrock-agentcore', region_name=REGION)
except Exception as e:
    print(f"Warning: Failed to create boto3 client for 'bedrock-agentcore'. {e}")
    print("Make sure your boto3 version supports AgentCore.")

# ãƒ¢ãƒ‡ãƒ«å‘¼ã³å‡ºã—ç”¨ (Boto3)
bedrock_runtime = boto3.client('bedrock-runtime', region_name=REGION)

ACTOR_ID = "engineer_tanaka_001" # ã‚¨ãƒ³ã‚¸ãƒ‹ã‚¢ç”°ä¸­ã•ã‚“
```

## 2. ãƒ¡ãƒ¢ãƒªãƒªã‚½ãƒ¼ã‚¹ä½œæˆé–¢æ•°
Episodeæ©Ÿèƒ½ã®ON/OFFã‚’åˆ‡ã‚Šæ›¿ãˆã¦ãƒ¡ãƒ¢ãƒªã‚’ä½œæˆã—ã¾ã™ã€‚

```python
def create_memory_resource(enable_episode=False):
    unique_id = str(uuid.uuid4())[:6]
    memory_name = f"trouble_mem_{unique_id}"
    
    # Semantic Memory (äº‹å®Ÿã®è¨˜æ†¶)
    strategies = [
        {
            "semanticMemoryStrategy": {
                "name": "factStrategy",
                "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}/facts"]
            }
        }
    ]
    
    # Episodeæ©Ÿèƒ½ã®ON/OFF
    if enable_episode:
        strategies.append({
            "episodicMemoryStrategy": {
                "name": "episodeStrategy",
                "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}/sessions/{sessionId}"],
                # Reflectionã¯Episodeã®è¦ªéšå±¤ã«è¨­å®š
                "reflectionConfiguration": {
                     "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}"]
                }
            }
        })
        label = "[Episode ON]"
    else:
        label = "[Episode OFF]"

    print(f"Creating Memory {label}: {memory_name} ...")
    memory_info = memory_client.create_memory_and_wait(
        name=memory_name,
        description=f"Troubleshooting memory {label}",
        strategies=strategies
    )
    # AgentCoreã®å ´åˆã€ã“ã®IDã¯KB ID(10æ¡)ã¨ã¯å½¢å¼ãŒç•°ãªã‚‹å ´åˆãŒã‚ã‚Šã¾ã™ãŒã€
    # AgentCoreç”¨ã®APIã‚’ä½¿ã†é™ã‚Šå•é¡Œã‚ã‚Šã¾ã›ã‚“ã€‚
    return memory_info['id']

# 2ã¤ã®ãƒ¡ãƒ¢ãƒªã‚’ä½œæˆ
mem_id_off = create_memory_resource(enable_episode=False)
mem_id_on = create_memory_resource(enable_episode=True)
```

## 3. éå»ã®æ—¥å ±ãƒ‡ãƒ¼ã‚¿ã®æŠ•å…¥

ã‚¨ãƒ³ã‚¸ãƒ‹ã‚¢ç”°ä¸­ã•ã‚“ã®ã€Œéå»ã®å¤±æ•—ï¼ˆç’°å¢ƒå¤‰æ•°å¿˜ã‚Œï¼‰ã€ã‚’ãƒ¡ãƒ¢ãƒªã«æ³¨å…¥ã—ã¾ã™ã€‚

```python
def inject_trouble_reports(memory_id):
    # --- Episode 1: éå»ã®å¤±æ•— ---
    session_id_1 = f"daily_report_20241001_{str(uuid.uuid4())[:4]}"
    messages_1 = [
        ("ä»Šæ—¥ã®æ—¥å ±ã§ã™ã€‚AWS Lambdaã®å®Ÿè£…ã§3æ™‚é–“ã‚‚ãƒãƒã£ã¦ã—ã¾ã„ã¾ã—ãŸã€‚", "USER"),
        ("ãã‚Œã¯å¤§å¤‰ã§ã—ãŸã­ã€‚ã©ã®ã‚ˆã†ãªã‚¨ãƒ©ãƒ¼ã§ã—ãŸã‹ï¼Ÿ", "ASSISTANT"),
        ("å®Ÿè¡Œæ™‚ã«ã‚­ãƒ¼ã‚¨ãƒ©ãƒ¼ãŒå‡ºã¦å‹•ã‹ãªã‹ã£ãŸã‚“ã§ã™ã€‚ã‚³ãƒ¼ãƒ‰ã¯åˆã£ã¦ã‚‹ã¯ãšãªã®ã«ã€‚", "USER"),
        ("åŸå› ã¯ç‰¹å®šã§ãã¾ã—ãŸã‹ï¼Ÿ", "ASSISTANT"),
        ("ã¯ã„ã€‚çµå±€ã€Lambdaã®ã‚³ãƒ³ã‚½ãƒ¼ãƒ«ç”»é¢ã§ç’°å¢ƒå¤‰æ•° `DB_CONNECTION_STRING` ã‚’è¨­å®šã—å¿˜ã‚Œã¦ãŸã®ãŒåŸå› ã§ã—ãŸã€‚", "USER"),
        ("ãªã‚‹ã»ã©ã€ç’°å¢ƒå¤‰æ•°ã®è¨­å®šæ¼ã‚Œã¯ã‚ˆãã‚ã‚‹ãƒŸã‚¹ã§ã™ã­ã€‚è¨˜éŒ²ã—ã¦ãŠãã¾ã™ã€‚", "ASSISTANT")
    ]
    
    print(f"Injecting Episode 1 (Env Var Failure) -> {memory_id} ...")
    memory_client.create_event(
        memory_id=memory_id,
        actor_id=ACTOR_ID,
        session_id=session_id_1,
        messages=messages_1
    )
    print("Done.")

print("--- Processing Episode OFF Memory ---")
inject_trouble_reports(mem_id_off)

print("\n--- Processing Episode ON Memory ---")
inject_trouble_reports(mem_id_on)
```

## 4. ã‚¨ãƒ¼ã‚¸ã‚§ãƒ³ãƒˆã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ (ä¿®æ­£ç‰ˆ)

**å¤‰æ›´ç‚¹:**
æ¤œç´¢APIã‚’ `retrieve` (KBç”¨) ã‹ã‚‰ `retrieve_memory_records` (AgentCoreç”¨) ã«å¤‰æ›´ã—ã¾ã—ãŸã€‚

```python
print("Waiting for AgentCore processing (sleeping 60s)...")
time.sleep(60)

def ask_agent_with_memory(memory_id, user_query, title):
    print(f"\n{'#'*60}")
    print(f" {title}")
    print(f"{'#'*60}")

    # 1. ãƒ¡ãƒ¢ãƒªã‹ã‚‰é–¢é€£æƒ…å ±ã‚’æ¤œç´¢ (Retrieve Memory Records)
    # AgentCoreå°‚ç”¨ã®APIã‚’ä½¿ç”¨ã—ã¾ã™
    try:
        retrieve_response = agentcore_runtime.retrieve_memory_records(
            memoryId=memory_id,
            searchCriteria={
                'searchQuery': user_query
                # å¿…è¦ã«å¿œã˜ã¦strategyNameãªã©ã®ãƒ•ã‚£ãƒ«ã‚¿ã‚’è¿½åŠ å¯èƒ½
            },
            limit=5
        )
        
        # ãƒ¬ã‚¹ãƒãƒ³ã‚¹æ§‹é€ ã®é•ã„ã«å¯¾å¿œ (recordsã‚­ãƒ¼ãªã©ã«å…¥ã£ã¦ã„ã‚‹æƒ³å®š)
        # å®Ÿéš›ã®ã‚­ãƒ¼åã¯SDKã®ãƒãƒ¼ã‚¸ãƒ§ãƒ³ã«ã‚ˆã‚Šã¾ã™ãŒã€'records' ã¾ãŸã¯ 'memoryRecords' ãŒä¸€èˆ¬çš„ã§ã™
        retrieved_results = retrieve_response.get('records', [])
        if not retrieved_results:
             retrieved_results = retrieve_response.get('memoryRecords', [])
        
        # æ¤œç´¢çµæœã‚’ãƒ†ã‚­ã‚¹ãƒˆåŒ–
        context_text = ""
        print("  [æ¤œç´¢ã•ã‚ŒãŸè¨˜æ†¶]")
        if not retrieved_results:
            print("   (ãªã—)")
            context_text = "(ç‰¹ã«é–¢é€£ã™ã‚‹è¨˜æ†¶ã¯ã‚ã‚Šã¾ã›ã‚“)"
        else:
            for res in retrieved_results:
                # çµæœã®æ§‹é€ ã‚‚KBã¨ã¯ç•°ãªã‚‹ãŸã‚ã€é©å®œèª¿æ•´
                # AgentCoreã§ã¯ 'content' ã ã‘ã§ãªããƒ¡ã‚¿ãƒ‡ãƒ¼ã‚¿ã‚‚å«ã¾ã‚Œã‚‹ã“ã¨ãŒå¤šã„
                content = res.get('content', {}).get('text', '')
                if not content: # æ§‹é€ ãŒé•ã†å ´åˆã®ãƒ•ã‚©ãƒ¼ãƒ«ãƒãƒƒã‚¯
                    content = str(res)
                
                print(f"   - {content[:100]}...")
                context_text += f"- {content}\n"

    except Exception as e:
        print(f"  Retrieve Error: {e}")
        context_text = "(è¨˜æ†¶ã®æ¤œç´¢ã«å¤±æ•—ã—ã¾ã—ãŸ)"

    # 2. ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆä½œæˆ
    prompt = f"""
ã‚ãªãŸã¯è¦ªåˆ‡ã§å„ªç§€ãªAIã‚¢ã‚·ã‚¹ã‚¿ãƒ³ãƒˆã§ã™ã€‚
ä»¥ä¸‹ã®ã€éå»ã®è¨˜æ†¶ã€‘ã‚’å‚è€ƒã«ã—ã¦ã€ãƒ¦ãƒ¼ã‚¶ãƒ¼ã®å›°ã‚Šã”ã¨ã«å¯¾ã—ã¦ã‚¢ãƒ‰ãƒã‚¤ã‚¹ã‚’ã—ã¦ãã ã•ã„ã€‚
ã‚‚ã—è¨˜æ†¶ã®ä¸­ã«ãƒ’ãƒ³ãƒˆãŒã‚ã‚Œã°ã€ãã‚Œã‚’å…·ä½“çš„ã«æŒ‡æ‘˜ã—ã¦ãã ã•ã„ã€‚

ã€éå»ã®è¨˜æ†¶ã€‘
{context_text}

ãƒ¦ãƒ¼ã‚¶ãƒ¼: {user_query}
AI:"""

    # 3. Amazon Nova Micro ã§å›ç­”ç”Ÿæˆ
    try:
        model_id = "amazon.nova-micro-v1:0"
        
        # Nova Microç”¨ã®ãƒªã‚¯ã‚¨ã‚¹ãƒˆãƒœãƒ‡ã‚£
        body = json.dumps({
            "inferenceConfig": {
                "max_new_tokens": 512
            },
            "messages": [
                {
                    "role": "user",
                    "content": [
                        {"text": prompt}
                    ]
                }
            ]
        })

        response = bedrock_runtime.invoke_model(
            modelId=model_id,
            body=body
        )
        
        response_body = json.loads(response.get('body').read())
        ai_response = response_body['output']['message']['content'][0]['text']
        
        print(f"\n  [ğŸ¤– AIã‚¨ãƒ¼ã‚¸ã‚§ãƒ³ãƒˆã®å›ç­” ({model_id})]")
        print(f"  {ai_response}")
        
    except Exception as e:
        print(f"  Invoke Error: {e}")


# æ¤œè¨¼å®Ÿè¡Œï¼
query = "LambdaãŒå‹•ã‹ãªã„ã‚“ã ã‚ˆã­ã€œãªã‚“ã§ã ã‚ã†"

ask_agent_with_memory(mem_id_off, query, "âŒ Episode OFF ã®å ´åˆ")
ask_agent_with_memory(mem_id_on, query, "âœ… Episode ON ã®å ´åˆ")
```

## æœŸå¾…ã•ã‚Œã‚‹çµæœï¼ˆLTã§ã®è¦‹ã›æ–¹ï¼‰

    * æ¤œç´¢çµæœ: ã€Œç’°å¢ƒå¤‰æ•°ã‚’å¿˜ã‚ŒãŸã€ã¨ã„ã†äº‹å®ŸãŒã†ã¾ãæ¤œç´¢ã•ã‚Œãªã„ã€ã¾ãŸã¯æ–­ç‰‡ã™ãã¦AIãŒæ–‡è„ˆã‚’ç†è§£ã§ããªã„ã€‚
    * **å›ç­”**: ã€Œãƒ­ã‚°ã‚’ç¢ºèªã—ã¾ã—ã‚‡ã†ã€ã€Œã‚³ãƒ¼ãƒ‰ã«ã‚¨ãƒ©ãƒ¼ã¯ã‚ã‚Šã¾ã›ã‚“ã‹ï¼Ÿã€ã¨ã„ã£ãŸ**ä¸€èˆ¬è«–**ã«ãªã‚‹ã¯ãšã§ã™ã€‚

2.  **Episode ONã®å ´åˆ**
    * æ¤œç´¢çµæœ: ã€Œéå»ã«ç’°å¢ƒå¤‰æ•°è¨­å®šæ¼ã‚Œã§å¤±æ•—ã—ãŸã€ã¨ã„ã†**ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰ï¼ˆè¦ç´„ï¼‰**ãŒãƒ’ãƒƒãƒˆã™ã‚‹ã€‚
    * **å›ç­”**: ã€Œä»¥å‰ã‚‚**ç’°å¢ƒå¤‰æ•°ã®è¨­å®šæ¼ã‚Œ**ã§LambdaãŒå‹•ã‹ãªã‹ã£ãŸã“ã¨ãŒã‚ã‚Šã¾ã—ãŸã­ã€‚ä»Šå›ã‚‚åŒã˜åŸå› ã§ã¯ã‚ã‚Šã¾ã›ã‚“ã‹ï¼Ÿç¢ºèªã—ã¦ã¿ã¦ãã ã•ã„ã€‚ã€ã¨ã„ã£ãŸ**ç¥å¯¾å¿œ**ã«ãªã‚‹ã¯ãšã§ã™ã€‚

```python
# ã‚¯ãƒªãƒ¼ãƒ³ã‚¢ãƒƒãƒ—
memory_client.delete_memory(memory_id=mem_id_off)
memory_client.delete_memory(memory_id=mem_id_on)
print("Cleanup complete.")
```
