```python
# 1. å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import uuid
import time
import json
from datetime import datetime

try:
    from bedrock_agentcore.memory import MemoryClient
    print("Successfully imported MemoryClient.")
except ImportError:
    # ç’°å¢ƒã«ã‚ˆã£ã¦ã¯ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ãŒå¿…è¦
    print("Installing bedrock-agentcore...")
    !pip install bedrock-agentcore strands-agents strands-agents-tools --quiet
    from bedrock_agentcore.memory import MemoryClient

# ã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆåˆæœŸåŒ–
REGION = "us-east-1" 
memory_client = MemoryClient(region_name=REGION)
ACTOR_ID = "engineer_tanaka_001" # ã‚¨ãƒ³ã‚¸ãƒ‹ã‚¢ç”°ä¸­ã•ã‚“
```

## 2. ãƒ¡ãƒ¢ãƒªãƒªã‚½ãƒ¼ã‚¹ä½œæˆé–¢æ•°
å‰å›åŒæ§˜ã€Episodeæ©Ÿèƒ½ã®ON/OFFã‚’åˆ‡ã‚Šæ›¿ãˆã¦ãƒ¡ãƒ¢ãƒªã‚’ä½œæˆã—ã¾ã™ã€‚
ï¼ˆãƒãƒ¼ãƒ ã‚¹ãƒšãƒ¼ã‚¹ã®ã‚¨ãƒ©ãƒ¼å¯¾ç­–æ¸ˆã¿ç‰ˆã§ã™ï¼‰

```python
def create_memory_resource(enable_episode=False):
    unique_id = str(uuid.uuid4())[:6]
    memory_name = f"trouble_mem_{unique_id}"
    
    # Semantic Memory (äº‹å®Ÿã®è¨˜æ†¶)
    strategies = [
        {
            "semanticMemoryStrategy": {
                "name": "factStrategy",
                "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}/facts"]
            }
        }
    ]
    
    # Episodeæ©Ÿèƒ½ã®ON/OFF
    if enable_episode:
        strategies.append({
            "episodicMemoryStrategy": {
                "name": "episodeStrategy",
                "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}/sessions/{sessionId}"],
                # Reflectionã¯Episodeã®è¦ªéšå±¤ã«è¨­å®š
                "reflectionConfiguration": {
                     "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}"]
                }
            }
        })
        label = "[Episode ON]"
    else:
        label = "[Episode OFF]"

    print(f"Creating Memory {label}: {memory_name} ...")
    memory_info = memory_client.create_memory_and_wait(
        name=memory_name,
        description=f"Troubleshooting memory {label}",
        strategies=strategies
    )
    return memory_info['id']

# 2ã¤ã®ãƒ¡ãƒ¢ãƒªã‚’ä½œæˆ
mem_id_off = create_memory_resource(enable_episode=False)
mem_id_on = create_memory_resource(enable_episode=True)
```

## 3. æ—¥å ±ãƒ‡ãƒ¼ã‚¿ã®æŠ•å…¥ï¼ˆã“ã“ãŒã‚·ãƒŠãƒªã‚ªã®è‚ï¼ï¼‰

ã‚¨ãƒ³ã‚¸ãƒ‹ã‚¢ç”°ä¸­ã•ã‚“ã®æ—¥å ±ãƒ‡ãƒ¼ã‚¿ã‚’æ³¨å…¥ã—ã¾ã™ã€‚
**Episode 1** ã§æ˜ç¢ºãªã€Œå¤±æ•—åŸå› ï¼ˆç’°å¢ƒå¤‰æ•°ï¼‰ã€ã‚’è¨˜éŒ²ã•ã›ã¾ã™ã€‚

```python
def inject_trouble_reports(memory_id):
    # --- Episode 1: éå»ã®å¤±æ•— ---
    session_id_1 = f"daily_report_20241001_{str(uuid.uuid4())[:4]}"
    messages_1 = [
        ("ä»Šæ—¥ã®æ—¥å ±ã§ã™ã€‚AWS Lambdaã®å®Ÿè£…ã§3æ™‚é–“ã‚‚ãƒãƒã£ã¦ã—ã¾ã„ã¾ã—ãŸã€‚", "USER"),
        ("ãã‚Œã¯å¤§å¤‰ã§ã—ãŸã­ã€‚ã©ã®ã‚ˆã†ãªã‚¨ãƒ©ãƒ¼ã§ã—ãŸã‹ï¼Ÿ", "ASSISTANT"),
        ("å®Ÿè¡Œæ™‚ã«ã‚­ãƒ¼ã‚¨ãƒ©ãƒ¼ãŒå‡ºã¦å‹•ã‹ãªã‹ã£ãŸã‚“ã§ã™ã€‚ã‚³ãƒ¼ãƒ‰ã¯åˆã£ã¦ã‚‹ã¯ãšãªã®ã«ã€‚", "USER"),
        ("åŸå› ã¯ç‰¹å®šã§ãã¾ã—ãŸã‹ï¼Ÿ", "ASSISTANT"),
        ("ã¯ã„ã€‚çµå±€ã€Lambdaã®ã‚³ãƒ³ã‚½ãƒ¼ãƒ«ç”»é¢ã§ç’°å¢ƒå¤‰æ•° `DB_CONNECTION_STRING` ã‚’è¨­å®šã—å¿˜ã‚Œã¦ãŸã®ãŒåŸå› ã§ã—ãŸã€‚", "USER"),
        ("ãªã‚‹ã»ã©ã€ç’°å¢ƒå¤‰æ•°ã®è¨­å®šæ¼ã‚Œã¯ã‚ˆãã‚ã‚‹ãƒŸã‚¹ã§ã™ã­ã€‚è¨˜éŒ²ã—ã¦ãŠãã¾ã™ã€‚", "ASSISTANT")
    ]
    
    print(f"Injecting Episode 1 (Env Var Failure) -> {memory_id} ...")
    memory_client.create_event(
        memory_id=memory_id,
        actor_id=ACTOR_ID,
        session_id=session_id_1,
        messages=messages_1
    )
    
    # ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰ã®åŒºåˆ‡ã‚Šã‚’ä½œã‚‹ãŸã‚ã®Wait
    time.sleep(2)

    # --- Episode 2: ç¾åœ¨ã®ç›¸è«‡ ---
    # â€»ã“ã“ã§ã¯ã€Œã¾ãŸå‹•ã‹ãªã„ã€ã¨è¨€ã†ã ã‘ã§ã€åŸå› ã¯è¨€ã£ã¦ã„ã¾ã›ã‚“
    session_id_2 = f"daily_report_20241015_{str(uuid.uuid4())[:4]}"
    messages_2 = [
        ("ã¾ãŸLambdaãŒå‹•ã‹ãªãã¦å›°ã£ã¦ã„ã¾ã™ã€‚", "USER"),
        ("çŠ¶æ³ã‚’æ•™ãˆã¦ã„ãŸã ã‘ã¾ã™ã‹ï¼Ÿ", "ASSISTANT"),
        ("å‰å›ã¨ã¯åˆ¥ã®é–¢æ•°ã‚’ä½œã£ã¦ã‚‹ã‚“ã§ã™ãŒã€ãƒ‡ãƒ—ãƒ­ã‚¤ã—ã¦ã‚‚å‹•ã‹ãªã„ã‚“ã§ã™ã€‚", "USER")
    ]
    
    print(f"Injecting Episode 2 (Trouble Again)   -> {memory_id} ...")
    memory_client.create_event(
        memory_id=memory_id,
        actor_id=ACTOR_ID,
        session_id=session_id_2,
        messages=messages_2
    )
    print("Done.")

print("--- Processing Episode OFF Memory ---")
inject_trouble_reports(mem_id_off)

print("\n--- Processing Episode ON Memory ---")
inject_trouble_reports(mem_id_on)
```

## 4. æ¤œç´¢ï¼†æ¯”è¼ƒæ¤œè¨¼

ã“ã“ã§ã‚¨ãƒ¼ã‚¸ã‚§ãƒ³ãƒˆã«ãªã£ãŸã¤ã‚‚ã‚Šã§æ¤œç´¢ã—ã¾ã™ã€‚
ãƒ¦ãƒ¼ã‚¶ãƒ¼ãŒ**ã€ŒLambda å‹•ã‹ãªã„ åŸå› ã€**ã¨èª¿ã¹ãŸæ™‚ã€ãƒ¡ãƒ¢ãƒªã‹ã‚‰ä½•ãŒå‡ºã¦ãã‚‹ã§ã—ã‚‡ã†ã‹ï¼Ÿ

å‡¦ç†å®Œäº†ã¾ã§ç´„60ç§’å¾…ã¡ã¾ã™...

```python
print("Waiting for AgentCore processing (sleeping 60s)...")
time.sleep(60)

def show_comparison_trouble(query):
    print(f"\n{'#'*60}")
    print(f" ğŸ” Query: ã€Œ{query}ã€")
    print(f"{'#'*60}\n")

    for title, mem_id in [("âŒ Episode OFF", mem_id_off), ("âœ… Episode ON", mem_id_on)]:
        print(f"[{title}]")
        try:
            results = memory_client.search_memory(
                memory_id=mem_id,
                actor_id=ACTOR_ID,
                text=query,
                limit=3
            )
            
            if not results:
                print("  (No results)")
                continue

            for i, res in enumerate(results, 1):
                # Episodeã®è¡¨ç¤ºãƒ­ã‚¸ãƒƒã‚¯
                if 'episodicMemory' in res:
                    ep = res['episodicMemory']
                    summary = ep.get('summary') or ep.get('text')
                    print(f"  {i}. ğŸ“˜ [ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰è¦ç´„] \"{summary}\"")
                    if 'insights' in ep and ep['insights']:
                        print(f"      ğŸ’¡ [æ´å¯Ÿ] {ep['insights']} (â†ã“ã‚ŒãŒé‡è¦ï¼)")
                else:
                    # Factã®è¡¨ç¤ºãƒ­ã‚¸ãƒƒã‚¯
                    content = res.get('text') or res.get('fact')
                    print(f"  {i}. ğŸ“„ [äº‹å®Ÿ] {content}")
        except Exception as e:
            print(f"  Error: {e}")
        print("\n" + "-"*30 + "\n")

# å®Ÿè¡Œï¼
show_comparison_trouble("Lambda å‹•ã‹ãªã„ åŸå› ")
```
