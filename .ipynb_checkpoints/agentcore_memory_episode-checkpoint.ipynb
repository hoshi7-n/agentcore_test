## 2. ã‚¤ãƒ³ãƒãƒ¼ãƒˆã¨ã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆåˆæœŸåŒ–
SDKã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆã—ã€MemoryClientã‚’åˆæœŸåŒ–ã—ã¾ã™ã€‚

```python
import uuid
import time
import json
from datetime import datetime

# æ­£ã—ã„ãƒ‘ãƒƒã‚±ãƒ¼ã‚¸åã§ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
try:
    from bedrock_agentcore.memory import MemoryClient
    print("Successfully imported MemoryClient.")
except ImportError:
    print("ERROR: bedrock_agentcore not found. Please check installation.")

# ã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆã®åˆæœŸåŒ– (ãƒªãƒ¼ã‚¸ãƒ§ãƒ³ã¯é©å®œå¤‰æ›´ã—ã¦ãã ã•ã„: us-east-1, us-west-2 ãªã©)
REGION = "us-east-1" 
memory_client = MemoryClient(region_name=REGION)

# æ¤œè¨¼ç”¨ã®ã‚¢ã‚¯ã‚¿ãƒ¼IDï¼ˆãƒ¦ãƒ¼ã‚¶ãƒ¼IDï¼‰
ACTOR_ID = "user_demo_001"
```

## 3. ãƒ¡ãƒ¢ãƒªä½œæˆé–¢æ•°ã®å®šç¾©

**Episodeæ©Ÿèƒ½ã®ON/OFF** ã‚’å®šç¾©ã—ã¾ã™ã€‚
`episodicMemoryStrategy` ã® `reflectionConfiguration` ã®ãƒãƒ¼ãƒ ã‚¹ãƒšãƒ¼ã‚¹ã‚’ä¿®æ­£ã—ã¦ã„ã¾ã™ã€‚

```python
def create_memory_resource(enable_episode=False):
    """
    æ¤œè¨¼ç”¨ã®ãƒ¡ãƒ¢ãƒªã‚’ä½œæˆã™ã‚‹é–¢æ•°
    """
    unique_id = str(uuid.uuid4())[:6]
    
    # ãƒã‚¤ãƒ•ãƒ³ã¯ä½¿ç”¨ä¸å¯ã®ãŸã‚ã€ã‚¢ãƒ³ãƒ€ãƒ¼ã‚¹ã‚³ã‚¢ã‚’ä½¿ç”¨
    memory_name = f"demo_mem_{unique_id}"
    
    # --- å…±é€šè¨­å®š: Semantic Memory (äº‹å®Ÿã®è¨˜æ†¶) ---
    strategies = [
        {
            "semanticMemoryStrategy": {
                "name": "factStrategy",
                "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}/facts"]
            }
        }
    ]
    
    # --- Episodeæ©Ÿèƒ½ã®ON/OFF ---
    if enable_episode:
        strategies.append({
            "episodicMemoryStrategy": {
                "name": "episodeStrategy",
                # ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰ï¼ˆå‡ºæ¥äº‹ã®è¦ç´„ï¼‰ã‚’ä¿å­˜ã™ã‚‹ãƒ‘ã‚¹
                "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}/sessions/{sessionId}"],
                # Reflectionã¯Episodeã®è¦ªéšå±¤ï¼ˆãƒ—ãƒ¬ãƒ•ã‚£ãƒƒã‚¯ã‚¹ï¼‰ã§ã‚ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™
                "reflectionConfiguration": {
                     "namespaces": ["/strategies/{memoryStrategyId}/actors/{actorId}"]
                }
            }
        })
        label = "[Episode ON]"
    else:
        label = "[Episode OFF]"

    print(f"Creating Memory {label}: {memory_name} ...")
    
    # ãƒ¡ãƒ¢ãƒªä½œæˆ
    memory_info = memory_client.create_memory_and_wait(
        name=memory_name,
        description=f"Demo memory {label}",
        strategies=strategies
    )
    
    print(f" -> Created ID: {memory_info['id']}")
    return memory_info['id']
```

## 4. æ¤œè¨¼ç”¨ãƒ¡ãƒ¢ãƒªã®ä½œæˆ
æ¯”è¼ƒã™ã‚‹ãŸã‚ã«2ã¤ã®ãƒ¡ãƒ¢ãƒªã‚’ä½œæˆã—ã¾ã™ã€‚

```python
# 1. Episodeæ©Ÿèƒ½ãªã—ï¼ˆå¾“æ¥ã®Semantic Memoryã®ã¿ï¼‰
memory_id_off = create_memory_resource(enable_episode=False)

# 2. Episodeæ©Ÿèƒ½ã‚ã‚Š
memory_id_on = create_memory_resource(enable_episode=True)
```

## 5. ä¼šè©±ãƒ‡ãƒ¼ã‚¿ã®æŠ•å…¥ï¼ˆã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ï¼‰

AgentCoreã® `create_event` ã‚’ä½¿ç”¨ã—ã¦ã€æ“¬ä¼¼çš„ã«ä¼šè©±å±¥æ­´ã‚’æ³¨å…¥ã—ã¾ã™ã€‚
ã“ã“ã§ã¯**ã€ŒåŒ—æµ·é“æ—…è¡Œã€**ã¨**ã€Œæ²–ç¸„æ—…è¡Œã€**ã¨ã„ã†2ã¤ã®ç•°ãªã‚‹ã‚»ãƒƒã‚·ãƒ§ãƒ³ã‚’æŠ•å…¥ã—ã¾ã™ã€‚

```python
def inject_conversation(memory_id):
    """
    æŒ‡å®šã•ã‚ŒãŸãƒ¡ãƒ¢ãƒªã«2ã¤ã®æ—…è¡Œè¨ˆç”»ã‚»ãƒƒã‚·ãƒ§ãƒ³ã‚’æ³¨å…¥ã™ã‚‹
    """
    # --- Session 1: åŒ—æµ·é“ ---
    session_id_1 = f"session_hokkaido_{str(uuid.uuid4())[:4]}"
    messages_1 = [
        ("å†¬ä¼‘ã¿ã®æ—…è¡Œã®è¨ˆç”»ã‚’ç«‹ã¦ãŸã„ã§ã™ã€‚", "USER"),
        ("ã„ã„ã§ã™ã­ã€‚ã©ã“ã«è¡ŒããŸã„ã§ã™ã‹ï¼Ÿ", "ASSISTANT"),
        ("åŒ—æµ·é“ã«è¡Œã£ã¦ã€é›ªæ™¯è‰²ã‚’è¦‹ãªãŒã‚‰æ¸©æ³‰ã«å…¥ã‚ŠãŸã„ã§ã™ã€‚", "USER"),
        ("æœ€é«˜ã§ã™ã­ã€‚é£Ÿäº‹ã®å¸Œæœ›ã¯ã‚ã‚Šã¾ã™ã‹ï¼Ÿ", "ASSISTANT"),
        ("ã‚«ãƒ‹ã‚’ã„ã£ã±ã„é£Ÿã¹ãŸã„ã§ã™ã€‚äºˆç®—ã¯10ä¸‡å††ä»¥å†…ã§ã€‚", "USER")
    ]
    
    print(f"Injecting Session 1 (Hokkaido) -> {memory_id} ...")
    memory_client.create_event(
        memory_id=memory_id,
        actor_id=ACTOR_ID,
        session_id=session_id_1,
        messages=messages_1
    )
    
    # å°‘ã—æ™‚é–“ã‚’ç©ºã‘ã‚‹ï¼ˆã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰åŒºåˆ‡ã‚Šã‚’æ˜ç¢ºã«ã™ã‚‹ãŸã‚ï¼‰
    time.sleep(2)

    # --- Session 2: æ²–ç¸„ ---
    session_id_2 = f"session_okinawa_{str(uuid.uuid4())[:4]}"
    messages_2 = [
        ("ã“ã‚“ã«ã¡ã¯ã€‚ä»Šåº¦ã¯å¤ä¼‘ã¿ã®æ—…è¡Œã§ã™ã€‚", "USER"),
        ("å¤ä¼‘ã¿ã§ã™ã­ã€‚ã”å¸Œæœ›ã¯ï¼Ÿ", "ASSISTANT"),
        ("æ²–ç¸„ã§ãƒ€ã‚¤ãƒ“ãƒ³ã‚°ãƒ©ã‚¤ã‚»ãƒ³ã‚¹ã‚’å–ã‚ŠãŸã„ã§ã™ã€‚", "USER"),
        ("ã‚¢ã‚¯ãƒ†ã‚£ãƒ–ã§ã™ã­ï¼é£Ÿäº‹ã¯ã©ã†ã—ã¾ã™ã‹ï¼Ÿ", "ASSISTANT"),
        ("ã‚½ãƒ¼ã‚­ãã°ãŒç¾å‘³ã—ã„åº—ã‚’æ•™ãˆã¦ã€‚äºˆç®—ã¯15ä¸‡å††ãã‚‰ã„ã§ã™ã€‚", "USER")
    ]
    
    print(f"Injecting Session 2 (Okinawa)  -> {memory_id} ...")
    memory_client.create_event(
        memory_id=memory_id,
        actor_id=ACTOR_ID,
        session_id=session_id_2,
        messages=messages_2
    )
    print("Done.")

# ä¸¡æ–¹ã®ãƒ¡ãƒ¢ãƒªã«ãƒ‡ãƒ¼ã‚¿ã‚’æŠ•å…¥
print("--- Processing Episode OFF Memory ---")
inject_conversation(memory_id_off)

print("\n--- Processing Episode ON Memory ---")
inject_conversation(memory_id_on)
```

## 6. çµæœç¢ºèª 1: è¨˜æ†¶ã®ä¸­èº«ï¼ˆãƒ€ãƒ³ãƒ—ï¼‰

ã¾ãšã¯å˜ç´”ã«ã€ãƒ¡ãƒ¢ãƒªå†…ã«ä¿å­˜ã•ã‚Œã¦ã„ã‚‹ãƒ‡ãƒ¼ã‚¿ä¸€è¦§ã‚’ç¢ºèªã—ã¾ã™ã€‚
â€» å‡¦ç†å®Œäº†ã¾ã§å°‘ã—æ™‚é–“ãŒã‹ã‹ã‚‹ãŸã‚å¾…æ©Ÿã—ã¾ã™ã€‚

```python
print("Waiting for AgentCore processing (sleeping 60s)...")
time.sleep(60)

def show_memory_dump(memory_id, title):
    print(f"\n{'='*10} {title} {'='*10}")
    try:
        facts = memory_client.list_facts(memory_id=memory_id, actor_id=ACTOR_ID)
        print(f"[ä¿å­˜ã•ã‚Œã¦ã„ã‚‹äº‹å®Ÿ] {len(facts)}å€‹")
        for f in facts[:3]: # å…ˆé ­3ã¤ã ã‘è¡¨ç¤º
            print(f" - {f.get('fact')}")
        if len(facts) > 3: print(" - ...")
    except: 
        print(" - Facts: ãªã—")

show_memory_dump(memory_id_off, "EPISODE OFF")
show_memory_dump(memory_id_on, "EPISODE ON")
```

## 7. æ¤œè¨¼: ã€Œã‚ã®æ™‚ã®è©±ã€è¦šãˆã¦ã‚‹ï¼Ÿã€

ã“ã“ãŒãƒ¡ã‚¤ãƒ³ã®æ¤œè¨¼ã§ã™ã€‚
**ã€Œã‚¨ãƒ¼ã‚¸ã‚§ãƒ³ãƒˆãŒéå»ã®æ–‡è„ˆã‚’ã©ã†æ€ã„å‡ºã™ã‹ã€** ã‚’ã€æ¤œç´¢ã‚¯ã‚¨ãƒªã‚’ä½¿ã£ã¦æ¯”è¼ƒã—ã¾ã™ã€‚

### æ¯”è¼ƒã™ã‚‹è³ªå•
1.  **ã€ŒåŒ—æµ·é“æ—…è¡Œã§ä½•ã‚’é£Ÿã¹ãŸã„ã¨è¨€ã£ãŸï¼Ÿã€**ï¼ˆå…·ä½“çš„ãªæ–‡è„ˆã®æ¤œç´¢ï¼‰
2.  **ã€Œå¤ã¨å†¬ã€ãã‚Œãã‚Œã®æ—…è¡Œè¨ˆç”»ã®é•ã„ã¯ï¼Ÿã€**ï¼ˆè¤‡æ•°ã®æ–‡è„ˆã‚’ã¾ãŸãæ¤œç´¢ï¼‰

```python
def compare_retrieval(query_text):
    print(f"\n{'='*60}")
    print(f"ğŸ” æ¤œè¨¼ã‚¯ã‚¨ãƒª: ã€Œ{query_text}ã€")
    print(f"{'='*60}")
    
    def perform_search(mem_id, label):
        print(f"\nğŸ‘‰ {label} ã®å ´åˆ:")
        try:
            # MemoryClientã®searchæ©Ÿèƒ½ã‚’ä½¿ã£ã¦é–¢é€£æƒ…å ±ã‚’æ¤œç´¢
            results = memory_client.search_memory(
                memory_id=mem_id,
                actor_id=ACTOR_ID,
                text=query_text,
                limit=3
            )
            
            if not results:
                print("   (ãƒ’ãƒƒãƒˆãªã—: é–¢é€£æƒ…å ±ãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã§ã—ãŸ)")
                return

            for i, res in enumerate(results):
                # çµæœãŒè¦‹ã‚„ã™ã„ã‚ˆã†ã«æ•´å½¢
                if 'fact' in res:
                    # äº‹å®Ÿï¼ˆFactï¼‰ã®å ´åˆ
                    print(f"   {i+1}. [äº‹å®Ÿ] {res['fact']}")
                elif 'episodicMemory' in str(res) or 'summary' in str(res).lower():
                    # ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰ï¼ˆEpisode/Summaryï¼‰ã®å ´åˆ
                    # SDKã®ãƒ¬ã‚¹ãƒãƒ³ã‚¹æ§‹é€ ã‹ã‚‰ãƒ†ã‚­ã‚¹ãƒˆéƒ¨åˆ†ã‚’æŠ½å‡ºã—ã¦è¡¨ç¤ºï¼ˆç°¡æ˜“çš„ãªå‡¦ç†ï¼‰
                    # å®Ÿéš›ã«ã¯ res['episodicMemory']['summary'] ãªã©ã‚’å‚ç…§
                    summary = str(res.get('summary', res))[:100] # é•·ã„ã®ã§ã‚«ãƒƒãƒˆ
                    print(f"   {i+1}. [æ–‡è„ˆ] {summary}...")
                else:
                    print(f"   {i+1}. [ãã®ä»–] {str(res)[:60]}...")

        except Exception as e:
            print(f"   Error: {e}")

    # æ¯”è¼ƒå®Ÿè¡Œ
    perform_search(memory_id_off, "Episode OFF (äº‹å®Ÿã®ã¿)")
    perform_search(memory_id_on,  "Episode ON  (æ–‡è„ˆã‚ã‚Š)")

# --- æ¤œè¨¼å®Ÿè¡Œ ---

# Q1: å…·ä½“çš„ãªæ–‡è„ˆã®ç¢ºèª
# OFF: ã€Œã‚«ãƒ‹ãŒå¥½ãã€ã¨ã„ã†äº‹å®Ÿã ã‘ãŒå‡ºã‚‹ï¼ˆæ–‡è„ˆãŒãªã„ï¼‰
# ON : ã€ŒåŒ—æµ·é“æ—…è¡Œã®è¨ˆç”»ã§ã€ã‚«ãƒ‹ã‚’é£Ÿã¹ãŸã„ã¨è©±ã—ãŸã€ã¨ã„ã†è¦ç´„ãŒå‡ºã‚‹
compare_retrieval("åŒ—æµ·é“æ—…è¡Œã§ã®é£Ÿäº‹ã®å¸Œæœ›ã¯ï¼Ÿ")

# Q2: å…¨ä½“åƒã®æ¯”è¼ƒ
# OFF: ãƒãƒ©ãƒãƒ©ã®äº‹å®ŸãŒå‡ºã‚‹ã ã‘ã§ã€å¤å†¬ã®åŒºåˆ¥ãŒã¤ã‹ãªã„å¯èƒ½æ€§ãŒã‚ã‚‹
# ON : å†¬ï¼ˆåŒ—æµ·é“ï¼‰ã¨å¤ï¼ˆæ²–ç¸„ï¼‰ã®ã‚¨ãƒ”ã‚½ãƒ¼ãƒ‰ãŒãã‚Œãã‚Œãƒ’ãƒƒãƒˆã™ã‚‹
compare_retrieval("å¤ã¨å†¬ã®æ—…è¡Œè¨ˆç”»ã®é•ã„ã‚’æ•™ãˆã¦")
```

```python
memory_client.delete_memory(memory_id=memory_id_off)
memory_client.delete_memory(memory_id=memory_id_on)
print("Cleanup complete.")
```
